```python
from haystack.document_stores import InMemoryDocumentStore #may not work for large deployment
document_store = InMemoryDocumentStore()
```

```python
from haystack.nodes import PDFToTextConverter ##https://haystack.deepset.ai/tutorials/preprocessing
```

```python
!wget --no-check-certificate https://dl.xpdfreader.com/xpdf-tools-linux-4.04.tar.gz && tar -xvf xpdf-tools-linux-4.04.tar.gz && sudo cp xpdf-tools-linux-4.04/bin64/pdftotext /usr/local/bin
```

```python
converter = PDFToTextConverter(remove_numeric_tables=True, valid_languages=["en"])
```

```python
doc_pdf = converter.convert(file_path="/home/jupyter/Covid19test.pdf", meta=None)[0]
```

```python
doc_pdf
```

```python
from haystack.nodes import PreProcessor
```

```python
preprocessor = PreProcessor(
    clean_empty_lines=True,
    clean_whitespace=True,
    clean_header_footer=False,
    split_by="word",
    split_length=100,
    split_respect_sentence_boundary=True,
)
```

```python
docs_default = preprocessor.process([doc_pdf])
```

```python
print(f"n_docs_input: 1\nn_docs_output: {len(docs_default)}")
```

```python
docs_default[0] #confirm data preprocessed correctly by looking at first "doc"
```

```python
document_store.write_documents(docs_default) #https://haystack.deepset.ai/components/document-store
```

```python
from haystack.nodes import QuestionGenerator, FARMReader 
#https://www.deepset.ai/blog/generate-questions-automatically-for-faster-annotation
#Note - copy/paste on my mac required replacing all " " - read in as different character
```

```python
question_generator = QuestionGenerator()
reader = FARMReader("deepset/roberta-base-squad2")
```

```python
from haystack.pipelines import QuestionAnswerGenerationPipeline #blog post has a typo, lists pipeline as singular
```

```python
qag_pipeline = QuestionAnswerGenerationPipeline(question_generator, reader)
```

```python
results = []
```

```python
for doc in document_store:
    results.append(qag_pipeline.run(documents=[doc]))
```
